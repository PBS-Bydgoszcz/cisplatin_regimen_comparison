```python
import pandas as pd
```

```python
df = pd.read_excel('all.xlsx')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(5, 6))
sns.boxplot(x='Treatment schedule', y='Cumulative Cisplatin dose (mg/m2)', data=df, color='skyblue')
plt.xlabel('Study arms')
plt.ylabel('Mean Cumulative Cisplatin Dose (mg/m2)')
plt.title('Cumulative Cisplatin Dose by Treatment Schedule')
plt.ylim(0, 400)
plt.xticks(rotation=45)
plt.show()

```
